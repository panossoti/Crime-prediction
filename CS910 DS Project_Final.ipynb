```python
# Use numpy to convert to arrays
# OFFENCE_DESCRIPTION_LABELS are the values we want to predict
OFFENCE_LABELS = np.array(cleaned_data_17_19['OFFENCE_DESCRIPTION_LABELS'])
# Remove the labels from the features
# axis 1 refers to the columns
# Saving feature names for later use
# Convert to numpy array
features= cleaned_data_17_19[['Year','Month','Week_Day','Hour','VICTIM_AGEGROUP_LABELS','VICTIM_RACE_LABELS','VICTIM_SEX_LABELS','OFFENCE_CATEGORY_LABELS']]
feature_list = list(features.columns)
```

```python
cleaned_data_17_19.head(5)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, OFFENCE_LABELS, test_size = 0.20, random_state = 42)


```

```python
#A Baseline a naive model would score 2.5% (1/42) lets see how we do

# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 45 decision trees
rf = RandomForestRegressor(n_estimators = 45, random_state = 42)
# Train the model on training data
rf.fit(train_features, train_labels);
```

```python
predictions = rf.predict(test_features)
```

```python
predictions=np.round(predictions)
```

```python
f1_score(predictions,test_labels,average='weighted')
```

```python
randomConfMatrix=confusion_matrix(predictions,test_labels)
```

```python
plt.figure(figsize=[24,24])

sn.heatmap(randomConfMatrix, annot=True)
plt.show()
```

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")


plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mutual_info_series)
```
